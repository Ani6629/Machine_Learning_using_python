```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split

my_data = pd.read_csv("Buy_Computer.csv", delimiter=",")
X = my_data[1:]
my_data = pd.get_dummies(my_data, columns=[ 'age', 'income','credit_rating'], drop_first= True)
my_data

my_data[0:14]
print(my_data)
print ("Dataset Length: ", len(my_data)) 
print ("Dataset Shape: ", my_data.shape) 
print ("Dataset: ",my_data.head()) 

feature_cols= ['student', 'Buy_Computer' , 'age_senior' ,'age_youth', 'income_low', 'income_medium', 'credit_rating_fair']
X = my_data[feature_cols]
Y = my_data.Buy_Computer
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)  
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree = drugTree.fit(X_train,y_train)
predTree = drugTree.predict(X_test)

from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))

from sklearn.externals.six import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
%matplotlib inline 

dot_data = StringIO()
filename = "Buy_Computer.png"
featureNames = feature_cols
targetNames = my_data["Buy_Computer"].unique().tolist()
out=tree.export_graphviz(drugTree,feature_names=featureNames, out_file=dot_data, class_names= np.unique(y_train), filled=True,  special_characters=True,rotate=False)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img,interpolation='nearest')



```
